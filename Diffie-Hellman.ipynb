Протокол Ди́ффи — Хе́ллмана (англ. Diffie–Hellman key exchange protocol, DH) — криптографический протокол, позволяющий двум и более сторонам получить общий секретный ключ, используя не защищенный от прослушивания канал связи. Полученный ключ используется для шифрования дальнейшего обмена с помощью алгоритмов симметричного шифрования.

```python
p = 97  # открытое простое число размерностью 2048, 3072, 4096 бит
g = 2   # g∈{1..p} первообразный корень по модулю p, зачастую g = 2 или 5
```

```python
a = 6  # a∈{1..p-1} Alice приватный ключ
# A = g**a % p
A = pow(g, a, p)  # Alice публичный ключ для Bob
print(f'🔓 Alice public key: {A}, send to Bob')
```

```python
b = 9  # b∈{1..p-1} Bob приватный ключ
B = pow(g, b, p)  # Bob публичный ключ для Alice
print(f'🔓 Bob public key: {B}, send to Alice')
```

Получение единого приватного ключа для симметричного шифрования

```python
a_shared = pow(B, a, p)
print(f"🔒 Alice shared private key: {a_shared}")
b_shared = pow(A, b, p)
print(f"🔒 Bob shared private key: {b_shared}")
```
