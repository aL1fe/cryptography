Эллиптическая кривая — это просто множество точек, описываемое уравнением: y^2 = x^3 + ax + b  

```python
import numpy as np
import matplotlib.pyplot as plt
```

Примеры эллиптических кривых при a = -3, -2, -1, 0, 1, 2; b = 1

```python
b = 1
x = np.linspace(-2, 5, 10_000)  # массив из 10_000 равномерно распределённых значений от -2 до 5

for a in range(2, -4, -1):
    plt.figure(figsize=(5, 5))
    plt.axhline(0, color='black', linewidth=0.7)   # ось X
    plt.axvline(0, color='black', linewidth=0.7)   # ось Y
    plt.grid(True)

    y = np.sqrt(x**3 + a*x + b)

    plt.plot(x, y)
    plt.plot(x, -y)

    if a > 0:
        plt.title(f"y^2 = x^3 + {a}x + {b}")
    elif a < 0:
        plt.title(f"y^2 = x^3 - {abs(a)}x + {b}")
    else:
        plt.title(f"y^2 = x^3 + {b}")

    plt.show()
```

```python
x = np.linspace(0, 5, 10_000)

plt.figure(figsize=(5, 5))
plt.axhline(0, color='black', linewidth=0.7)   # ось X
plt.axvline(0, color='black', linewidth=0.7)   # ось Y
plt.grid(True)

y = np.sqrt(x**3)

plt.plot(x, y)
plt.plot(x, -y)

plt.title("Кривая с точкой возврата (каспом): y^2 = x^3")
plt.show()
```

```python
x = np.linspace(-2, 5, 10_000)

plt.figure(figsize=(5, 5))
plt.axhline(0, color='black', linewidth=0.7)   # ось X
plt.axvline(0, color='black', linewidth=0.7)   # ось Y
plt.grid(True)

y = np.sqrt(x**3 - 3 * x + 2)

plt.plot(x, y)
plt.plot(x, -y)

plt.title("Кривая с самопересечением: y^2 = x^3 - 3x + 2")
plt.show()
```
