Пример шифровки и расшифровки сообщений

```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP


# Генерация ключей для Алисы и Боба
alice_key = RSA.generate(1024)
bob_key = RSA.generate(1024)

# Алиса и Боб обмениваются клчами друг с другом
print(f"Привет, Боб! Вот мой публичный ключ '{alice_key.publickey()}'. Алиса.")
print(f"Привет, Алиса! Вот мой публичный ключ '{bob_key.publickey()}'. Боб.")
print("-" * 32)

# Алиса шифрованное сообщение Бобу его публичным ключом
alice_message = "Hello Bob, this is my message. Alice.".encode("utf-8")
bob_public_cipher = PKCS1_OAEP.new(bob_key.publickey())  # Используем публичный ключ Боба для шифрования
alice_encrypted_msg = bob_public_cipher.encrypt(alice_message) # Зашифрованное сообщение
print("Алиса зашифровала и отправила сообщение Бобу.")
print(f"Текст зашифрованного сообщения: {alice_message}")
print(f"Зашифрованное сообщение: {alice_encrypted_msg}")
print("-" * 32)

# Боб расшифровывает сообщение своим приватным ключом
cipher_bob = PKCS1_OAEP.new(bob_key)
decrypted_msg = cipher_bob.decrypt(alice_encrypted_msg)
print("Боб получил и расшифровал сообщение:")
print(decrypted_msg)
print("-" * 32)

# Боб отвечает Алисе
bob_message = "Hello Alice, I got your message and decripted it. Bob.".encode("utf-8")
alice_public_cipher = PKCS1_OAEP.new(alice_key.publickey())
bob_encrypted_reply = alice_public_cipher.encrypt(bob_message)
print("Боб зашифровал ответ и отправил Алисе.")
print(f"Текст зашифрованного сообщения: {bob_message}")
print(f"Зашифрованное сообщение: {bob_encrypted_reply}")
print("-" * 32)

# Алиса расшифровывает ответ своим приватным ключом
cipher_alice = PKCS1_OAEP.new(alice_key)
decrypted_reply = cipher_alice.decrypt(bob_encrypted_reply)
print("Алиса получила и расшифровала сообщение:")
print(decrypted_reply)
```

Пример подписи сообщений и проверки подписи

```python
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256


# Генерация ключей для Алисы и Боба
alice_key = RSA.generate(1024)
bob_key = RSA.generate(1024)

# Алиса и Боб обмениваются публичными ключами
print(f"Привет, Боб! Вот мой публичный ключ: {alice_key.publickey()}. Алиса.")
print(f"Привет, Алиса! Вот мой публичный ключ: {bob_key.publickey()}. Боб.")
print("-" * 32)

# 1. Алиса создаёт сообщение и подписывает его
message_from_alice = "Это сообщение от Алисы.".encode("utf-8")
hash_alice = SHA256.new(message_from_alice)
signature_alice = pkcs1_15.new(alice_key).sign(hash_alice)
print("Алиса подписала сообщение и отправила его Бобу.")
print(f"Сообщение: {message_from_alice.decode('utf-8')}")
print(f"Подпись: {signature_alice}")
print("-" * 32)

# 2. Боб проверяет подпись Алисы
try:
    hash_alice_verify = SHA256.new(message_from_alice)
    pkcs1_15.new(alice_key.publickey()).verify(hash_alice_verify, signature_alice)
    print("Боб проверил подпись: Сообщение подлинное и не изменено!")
except (ValueError, TypeError):
    print("Боб: Подпись недействительна!")
print("-" * 32)

# 3. Боб отвечает Алисе, подписывая своё сообщение
message_from_bob = "Алиса, получил твоё сообщение.".encode("utf-8")
hash_bob = SHA256.new(message_from_bob)
signature_bob = pkcs1_15.new(bob_key).sign(hash_bob)
print("Боб подписал ответ и отправил Алисе.")
print(f"Сообщение: {message_from_bob.decode('utf-8')}")
print(f"Подпись: {signature_bob}")
print("-" * 32)

# 4. Алиса проверяет подпись Боба
try:
    hash_bob_verify = SHA256.new(message_from_bob)
    pkcs1_15.new(bob_key.publickey()).verify(hash_bob_verify, signature_bob)
    print("Алиса проверила подпись: Сообщение подлинное и не изменено!")
except (ValueError, TypeError):
    print("Алиса: Подпись недействительна!")
```
